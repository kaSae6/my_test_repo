```python
import pandas as pd
import matplotlib.pyplot as plt

umsatz = pd.read_csv('https://raw.githubusercontent.com/opencampus-sh/einfuehrung-in-data-science-und-ml/main/umsatzdaten_gekuerzt.csv')
kiwo = pd.read_csv('https://raw.githubusercontent.com/opencampus-sh/einfuehrung-in-data-science-und-ml/main/kiwo.csv')
weather = pd.read_csv('https://raw.githubusercontent.com/opencampus-sh/einfuehrung-in-data-science-und-ml/main/wetter.csv')

```

```python
fig, ax = plt.subplots(1, 3)

ax[0].bar(weather['Datum'], weather['Temperatur'])
ax[1].scatter(weather['Datum'], weather['Temperatur'])

for i in range(3):
    ax[i].set_xlabel('Date')
    ax[i].set_ylabel('Temperature [deg C]')
```
